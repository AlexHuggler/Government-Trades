# Nancy Pelosi trades from Capitol Trades
This notebook pulls Nancy Pelosi's trade disclosures from [Capitol Trades](https://www.capitoltrades.com/) and aggregates them by owner and transaction type.

```python
# Install dependencies (safe to run multiple times)
%pip install -q -r requirements.txt
```

```python
from scrape_trades import scrape_pelosi_trades, aggregate_trades, ColumnHints

# Adjust max_pages if you want to pull more than the default 3 pages
raw_trades = scrape_pelosi_trades(max_pages=3)
raw_trades.head()
```

```python
summary = aggregate_trades(raw_trades, ColumnHints(transaction=None, owner=None))
summary
```

```python
raw_path = 'pelosi_raw_trades.csv'
agg_path = 'pelosi_aggregated_trades.csv'
raw_trades.to_csv(raw_path, index=False)
summary.to_csv(agg_path, index=False)
print(f'Saved raw data to {raw_path}')
print(f'Saved aggregated summary to {agg_path}')
```
